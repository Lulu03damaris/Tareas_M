```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funci贸n para graficar el sistema (a)
def plot_system_a():
    x = np.linspace(-10, 10, 400)
    y1 = -0.5 * x
    y2 = x

    plt.figure()
    plt.plot(x, y1, label='$x_1 + 2x_2 = 0$')
    plt.plot(x, y2, label='$x_1 - x_2 = 0$')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.title('Sistema (a)')
    plt.show()

# Funci贸n para graficar el sistema (b)
def plot_system_b():
    x = np.linspace(-10, 10, 400)
    y1 = 1.5 - 0.5 * x
    y2 = -1.5 - 0.5 * x

    plt.figure()
    plt.plot(x, y1, label='$x_1 + 2x_2 = 3$')
    plt.plot(x, y2, label='$-2x_1 - 4x_2 = 6$')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.title('Sistema (b)')
    plt.show()

# Funci贸n para graficar el sistema (c)
def plot_system_c():
    x = np.linspace(-10, 10, 400)
    y1 = -2 - 2 * x
    y2 = 2 - x
    y3 = (5 - x) / 3

    plt.figure()
    plt.plot(x, y1, label='$2x_1 + x_2 = -1$')
    plt.plot(x, y2, label='$x_1 + x_2 = 2$')
    plt.plot(x, y3, label='$x_1 - 3x_2 = 5$')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.title('Sistema (c)')
    plt.show()

# Funci贸n para graficar el sistema (d)
def plot_system_d():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Define planos
    x = np.linspace(-10, 10, 400)
    y = np.linspace(-10, 10, 400)
    X, Y = np.meshgrid(x, y)

    Z1 = 1 - 2*X - Y
    Z2 = -1 - 2*X - 4*Y

    ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100)
    ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100)

    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_zlabel('$x_3$')

    plt.title('Sistema (d)')
    plt.show()

# Llamar a las funciones para graficar los sistemas
plot_system_a()
plot_system_b()
plot_system_c()
plot_system_d()
```

```python
import numpy as np


def gauss_elimination_with_rounding(A, b):
    n = len(A)
    M = A

    for i in range(n):
        M[i] = [round(elem, 2) for elem in M[i]]
        b[i] = round(b[i], 2)

    for k in range(n):
        for i in range(k+1, n):
            if M[i][k] != 0.0:
                lam = round(M[i][k] / M[k][k], 2)
                M[i] = [round(M[i][j] - lam * M[k][j], 2) for j in range(n)]
                b[i] = round(b[i] - lam * b[k], 2)

    x = [0 for i in range(n)]
    for i in range(n-1, -1, -1):
        s = sum(M[i][j] * x[j] for j in range(i, n))
        x[i] = round((b[i] - s) / M[i][i], 2)
    return x

# Sistema a)
A1 = np.array([
    [-1, 4, 1],
    [5/3, 2/3, 2/3],
    [2, 1, 4]
])
b1 = np.array([8, 1, 11])

# Sistema b)
A2 = np.array([
    [4, 2, -1],
    [1/9, 1/2, -1/3],
    [1, 4, 2]
])
b2 = np.array([-5, -1, 9])


x1 = gauss_elimination_with_rounding(A1, b1)
x2 = gauss_elimination_with_rounding(A2, b2)


print("Solucion del sistema (a):", x1)
print("\nSolucion del sistema (b):", x2)
```

```python
import numpy as np
import logging

logging.basicConfig(level=logging.INFO)

def eliminacion_gaussiana(A: np.ndarray) -> np.ndarray:
    if not isinstance(A, np.ndarray):
        logging.debug("Convirtiendo A a numpy array.")
        A = np.array(A, dtype=float)  # convertir en float, porque si no, convierte en enteros
    assert A.shape[0] == A.shape[1] - 1, "La matriz B debe ser de tama帽o n-by-(n+1)."
    n = A.shape[0]
    n_adds = 0
    n_mults = 0
    intercambios = []
    for i in range(0, n - 1):  # loop por columna

        # --- encontrar pivote
        p = None  # default, first element
        for pi in range(i, n):
            if A[pi, i] == 0:
                # must be nonzero
                continue

            if p is None:
                # first nonzero element
                p = pi
                continue

            if abs(A[pi, i]) > abs(A[p, i]):
                p = pi

        if p is None:
            # no pivot found.
            logging.info(f"\n{A}")
            raise ValueError("No existe soluci贸n 煤nica.")

        if p != i:
            logging.info(f"Intercambiando filas {i} y {p}.")
            # swap rows
            A[[i, p]] = A[[p, i]]
            intercambios.append((i, p))

        # --- Eliminaci贸n: loop por fila
        for j in range(i + 1, n):
            m = A[j, i] / A[i, i]
            A[j, i:] -= m * A[i, i:]
            n_mults += 1 + (n + 1 - i - 1)
            n_adds += n + 1 - i - 1

        logging.info(f"\n{A}")

    if A[n - 1, n - 1] == 0:
        if A[n - 1, n] == 0:
            raise ValueError("Infinitas soluciones.")
        else:
            raise ValueError("Sin solucion",print("Sin soluci贸n."))

    # --- Sustituci贸n hacia atr谩s
    solucion = np.zeros(n)
    solucion[n - 1] = A[n - 1, n] / A[n - 1, n - 1]
    n_mults += 1

    for i in range(n - 2, -1, -1):
        suma = 0
        n_adds -= 1
        for j in range(i + 1, n):
            suma += A[i, j] * solucion[j]
            n_mults += 1
            n_adds += 1
        solucion[i] = (A[i, n] - suma) / A[i, i]
        n_adds += 1
        n_mults += 1

    return solucion, n_adds, n_mults, intercambios
```

```python
import numpy as np

def gaussian_elimination_32bit(A, b):
    A = A.astype(np.float32)
    b = b.astype(np.float32)

    n = len(b)

    # Forward elimination
    for i in range(n):
        # Pivot for maximum value in column i to avoid numerical instability
        max_row = np.argmax(np.abs(A[i:n, i])) + i
        A[[i, max_row]] = A[[max_row, i]]
        b[[i, max_row]] = b[[max_row, i]]

        # Make the diagonal contain all ones
        pivot = A[i, i]
        A[i] = A[i] / pivot
        b[i] = b[i] / pivot

        # Make the elements below the pivot in column i equal to zero
        for j in range(i + 1, n):
            factor = A[j, i]
            A[j] -= factor * A[i]
            b[j] -= factor * b[i]

    # Back substitution
    x = np.zeros(n, dtype=np.float32)
    for i in range(n - 1, -1, -1):
        x[i] = b[i] - np.dot(A[i, i + 1:], x[i + 1:])

    return x

# Definici贸n de los sistemas de ecuaciones
systems = [
    (np.array([[1/4, 1/5, 1/6], [1/3, 1/4, 1/5], [1/2, 1, 2]], dtype=np.float32), np.array([9, 8, 8], dtype=np.float32)),
    (np.array([[3.333, 15920, -10.333], [2.222, 16.71, 9.612], [1.5611, 5.1791, 1.6852]], dtype=np.float32), np.array([15913, 28.544, 8.4254], dtype=np.float32)),
    (np.array([[1, 1/2, 1/3, 1/4], [1/2, 1/3, 1/4, 1/5], [1/3, 1/4, 1/5, 1/6], [1/4, 1/5, 1/6, 1/7]], dtype=np.float32), np.array([1/6, 1/7, 1/8, 1/9], dtype=np.float32)),
    (np.array([[2, 1, -1, 1, -3], [1, 0, 2, -1, 1], [0, -2, -1, 1, -5], [3, 1, -4, 0, 5], [1, -1, -1, -1, 1]], dtype=np.float32), np.array([7, 2, -5, 6, -3], dtype=np.float32))
]

# Resolver cada sistema
solutions = [gaussian_elimination_32bit(A, b) for A, b in systems]
for i, sol in enumerate(solutions):
    print(f'Sistema {i+1}: {sol}')
```

```python
def analyze_system_with_alpha():
    from sympy import symbols, Eq, solve

    alpha = symbols('alpha')
    x1, x2, x3 = symbols('x1 x2 x3')

    equations = [
        Eq(x1 - x2 + alpha * x3, -2),
        Eq(-x1 + 2 * x2 - alpha * x3, 3),
        Eq(alpha * x1 + x2 + x3, 2)
    ]

    sol = solve(equations, (x1, x2, x3, alpha), dict=True)
    alpha_values = [s[alpha] for s in sol]

    for val in alpha_values:
        if val != alpha:
            print(f"Para 伪 = {val}, el sistema tiene soluci贸n.")
        else:
            print("Para valores de 伪 distintos a los encontrados, el sistema no tiene soluci贸n.")

analyze_system_with_alpha()
```

```python
def system_of_species():
    A = np.array([
        [1, 2, 0, 3],
        [1, 0, 2, 2],
        [0, 0, 1, 1]
    ])
    x = np.array([1000, 500, 350, 400])
    b = np.array([3500, 2700, 900])

    # Verificar si hay suficiente alimento
    suficiente_alimento = np.all(np.dot(A, x) <= b)
    print(f"驴Hay suficiente alimento? {'S铆' if suficiente_alimento else 'No'}")

    # N煤mero m谩ximo de animales que se podr铆a agregar
    incrementos_maximos = b - np.dot(A, x)
    print(f"Incrementos m谩ximos posibles: {incrementos_maximos}")

    # Incrementos si la especie 1 se extingue
    A_ext1 = A[:, 1:]
    incrementos_ext1 = b - np.dot(A_ext1, x[1:])
    print(f"Incrementos posibles si la especie 1 se extingue: {incrementos_ext1}")

    # Incrementos si la especie 2 se extingue
    A_ext2 = np.delete(A, 1, axis=1)
    incrementos_ext2 = b - np.dot(A_ext2, np.delete(x, 1))
    print(f"Incrementos posibles si la especie 2 se extingue: {incrementos_ext2}")

system_of_species()
```

```python
def gauss_jordan(A, b):
    n = len(b)
    A = A.astype(float)
    b = b.astype(float).reshape(-1, 1)
    augmented_matrix = np.hstack((A, b))

    for i in range(n):
        augmented_matrix[i] = augmented_matrix[i] / augmented_matrix[i, i]
        for j in range(n):
            if i != j:
                augmented_matrix[j] -= augmented_matrix[i] * augmented_matrix[j, i]

    return augmented_matrix[:, -1]

# Resolver cada sistema con Gauss-Jordan
solutions_gj = [gauss_jordan(A, b) for A, b in systems]
for i, sol in enumerate(solutions_gj):
    print(f'Sistema {i+1} (Gauss-Jordan): {sol}')
```
