```python
#Ejercicio 1_a
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Datos
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Ajuste polinomial de grado 1
coeffs = np.polyfit(x, y, 1)
p1 = np.poly1d(coeffs)
y_pred1 = p1(x)
error1 = mean_squared_error(y, y_pred1)

# Graficar
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='coral', label='Datos')
x_fit = np.linspace(min(x), max(x), 100)
plt.plot(x_fit, p1(x_fit), label=f'Ajuste lineal (grado 1), MSE: {error1:.2f}\nCoef: {coeffs}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste lineal')
plt.show()
```

```python
#Ejercicio 1_d
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Datos
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

log_y = np.log(y)
coeffs = np.polyfit(x, log_y, 1)
a_exp, log_b = coeffs
b_exp = np.exp(log_b)
y_pred_exp = b_exp * np.exp(a_exp * x)
error_exp = mean_squared_error(y, y_pred_exp)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='brown', label='Datos')
x_fit = np.linspace(min(x), max(x), 100)
y_exp_fit = b_exp * np.exp(a_exp * x_fit)
plt.plot(x_fit, y_exp_fit, label=f'Ajuste exponencial (be^ax), MSE: {error_exp:.2f}\nCoef: a={a_exp:.4f}, b={b_exp:.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste exponencial')
plt.show()
```

```python
#Ejercicio 1_e
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Datos
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Ajuste de ley de potencia
log_x = np.log(x)
log_y = np.log(y)
coeffs = np.polyfit(log_x, log_y, 1)
a_pow, log_b = coeffs
b_pow = np.exp(log_b)
y_pred_pow = b_pow * x ** a_pow
error_pow = mean_squared_error(y, y_pred_pow)

# Graficar
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='hotpink', label='Datos')
x_fit = np.linspace(min(x), max(x), 100)
y_pow_fit = b_pow * x_fit ** a_pow
plt.plot(x_fit, y_pow_fit, label=f'Ajuste de ley de potencia (bx^a), MSE: {error_pow:.2f}\nCoef: a={a_pow:.4f}, b={b_pow:.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste de ley de potencia')
plt.show()
```
