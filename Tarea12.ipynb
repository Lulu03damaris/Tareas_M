```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de las soluciones reales
def solution_a(t):
    return (1/5) * t * np.exp(3*t) - (1/25) * np.exp(3*t) + (1/25) * np.exp(-2*t)

def solution_b(t):
    return t + 1/(1-t)

def solution_c(t):
    return t * np.log(t) + 2*t

def solution_d(t):
    return (1/2) * np.sin(2*t) - (1/3) * np.cos(3*t) + (4/3)

# Definición de la función diferencial para cada ecuación
def f_a(t, y):
    return t * np.exp(3*t)

def f_b(t, y):
    return (1/(1-t))**2

def f_c(t, y):
    return np.log(t) + 2

def f_d(t, y):
    return np.cos(2*t) + np.sin(3*t)

# Método de Euler
def euler_method(f, t0, y0, h, n):
    t_values = [t0]
    y_values = [y0]
    for i in range(n):
        y0 += h * f(t0, y0)
        t0 += h
        t_values.append(t0)
        y_values.append(y0)
    return t_values, y_values

# Parámetros iniciales
t0 = 0
y0_a = solution_a(t0)
y0_b = solution_b(t0)
y0_c = solution_c(t0)
y0_d = solution_d(t0)

h = 0.1  # Tamaño de paso
n = 50   # Número de pasos

# Aplicación del método de Euler a cada ecuación
t_values, y_euler_a = euler_method(f_a, t0, y0_a, h, n)
_, y_euler_b = euler_method(f_b, t0, y0_b, h, n)
_, y_euler_c = euler_method(f_c, t0, y0_c, h, n)
_, y_euler_d = euler_method(f_d, t0, y0_d, h, n)

# Cálculo de las soluciones reales
y_real_a = [solution_a(t) for t in t_values]
y_real_b = [solution_b(t) for t in t_values]
y_real_c = [solution_c(t) for t in t_values]
y_real_d = [solution_d(t) for t in t_values]

# Cálculo del error
error_a = np.abs(np.array(y_real_a) - np.array(y_euler_a))
error_b = np.abs(np.array(y_real_b) - np.array(y_euler_b))
error_c = np.abs(np.array(y_real_c) - np.array(y_euler_c))
error_d = np.abs(np.array(y_real_d) - np.array(y_euler_d))

# Gráficos
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t_values, y_real_a, 'g', label='Solución Real')
plt.plot(t_values, y_euler_a, 'b--', label='Euler')
plt.plot(t_values, error_a, 'r', label='Error')
plt.title('a) $y(t) = \\frac{1}{5} t e^{3t} - \\frac{1}{25} e^{3t} + \\frac{1}{25} e^{-2t}$')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(t_values, y_real_b, 'g', label='Solución Real')
plt.plot(t_values, y_euler_b, 'b--', label='Euler')
plt.plot(t_values, error_b, 'r', label='Error')
plt.title('b) $y(t) = t + \\frac{1}{1-t}$')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(t_values, y_real_c, 'g', label='Solución Real')
plt.plot(t_values, y_euler_c, 'b--', label='Euler')
plt.plot(t_values, error_c, 'r', label='Error')
plt.title('c) $y(t) = t \\ln t + 2t$')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(t_values, y_real_d, 'g', label='Solución Real')
plt.plot(t_values, y_euler_d, 'b--', label='Euler')
plt.plot(t_values, error_d, 'r', label='Error')
plt.title('d) $y(t) = \\frac{1}{2} \\sin 2t - \\frac{1}{3} \\cos 3t + \\frac{4}{3}$')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from math import exp, log, sin, cos, pi

def exact_a(t):
    return (1 / 5) * t * exp(3 * t) - (1 / 25) * exp(3 * t) + (1 / 25) * exp(-2 * t)

def exact_b(t):
    return t + 1 / (1 - t)

def exact_c(t):
    return t * log(t) + 2 * t

def exact_d(t):
    return 1/2 * sin(2 * t) - 1/3 * cos(3 * t) + 4/3

def problem_a(t, y):
    return t * exp(3 * t) - 2 * y

def problem_b(t, y):
    return 1 + (t - y)**2

def problem_c(t, y):
    return 1 + y/t

def problem_d(t, y):
    return cos(2 * t) + sin(3 * t)

# Método de Euler
def ODE_euler(a, b, f, y_t0, h):
    n_steps = int((b - a) / h) + 1
    t_values = np.linspace(a, b, n_steps)
    y_values = np.zeros(n_steps)
    y_values[0] = y_t0

    for i in range(1, n_steps):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return y_values, t_values

h = 0.1  # tamaño del paso

# Problema (a)
print("Resultados del problema (a):")
ys_a, ts_a = ODE_euler(a=0.0, b=1.0, f=problem_a, y_t0=0.0, h=h)
for t, y in zip(ts_a, ys_a):
    exact_y = exact_a(t)
    error_real = abs(exact_y - y)
    print(f"t = {t:.2f}, y = {y:.4f}, exacta = {exact_y:.4f}, error real = {error_real:.4f}")

# Problema (b)
print("\nResultados del problema (b):")
ys_b, ts_b = ODE_euler(a=2.0, b=3.0, f=problem_b, y_t0=1.0, h=h)
for t, y in zip(ts_b, ys_b):
    exact_y = exact_b(t)
    error_real = abs(exact_y - y)
    print(f"t = {t:.2f}, y = {y:.4f}, exacta = {exact_y:.4f}, error real = {error_real:.4f}")

# Problema (c)
print("\nResultados del problema (c):")
ys_c, ts_c = ODE_euler(a=1.0, b=2.0, f=problem_c, y_t0=2.0, h=h)
for t, y in zip(ts_c, ys_c):
    exact_y = exact_c(t)
    error_real = abs(exact_y - y)
    print(f"t = {t:.2f}, y = {y:.4f}, exacta = {exact_y:.4f}, error real = {error_real:.4f}")

# Problema (d)
print("\nResultados del problema (d):")
ys_d, ts_d = ODE_euler(a=0.0, b=pi, f=problem_d, y_t0=1.0, h=h)
for t, y in zip(ts_d, ys_d):
    exact_y = exact_d(t)
    error_real = abs(exact_y - y)
    print(f"t = {t:.2f}, y = {y:.4f}, exacta = {exact_y:.4f}, error real = {error_real:.4f}")
```

```python
import numpy as np

# Definir las funciones para cada problema
def problem_a(t, y):
    return t * np.exp(3 * t) - 2 * y

def problem_b(t, y):
    return 1 + (t - y)**2

def problem_c(t, y):
    return 1 + y / t

def problem_d(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

# Derivadas segundas para el método de Taylor de orden 2
def problem_a_second(t, y, yp):
    return 3 * t * np.exp(3 * t) - 2 * yp

def problem_b_second(t, y, yp):
    return -2 * yp * (t - y)

def problem_c_second(t, y, yp):
    return -yp / t

def problem_d_second(t, y, yp):
    return -2 * np.sin(2 * t) + 3 * np.cos(3 * t)

# Método de Taylor de orden 2
def taylor_order2(a, b, f, f_second, y_t0, h):
    n_steps = int((b - a) / h) + 1
    t_values = np.linspace(a, b, n_steps)
    y_values = np.zeros(n_steps)
    y_values[0] = y_t0

    for i in range(1, n_steps):
        t = t_values[i - 1]
        y = y_values[i - 1]
        yp = f(t, y)
        ypp = f_second(t, y, yp)
        y_values[i] = y + h * yp + (h**2 / 2) * ypp

    return y_values, t_values

# Parámetros y ejecución para cada problema
# Problema (a)
ys_a, ts_a = taylor_order2(a=0.0, b=1.0, f=problem_a, f_second=problem_a_second, y_t0=0.0, h=0.5)
print("Resultados para el problema (a):")
for t, y in zip(ts_a, ys_a):
    print(f"t = {t:.2f}, y = {y:.4f}")

# Problema (b)
ys_b, ts_b = taylor_order2(a=2.0, b=3.0, f=problem_b, f_second=problem_b_second, y_t0=1.0, h=0.5)
print("\nResultados para el problema (b):")
for t, y in zip(ts_b, ys_b):
    print(f"t = {t:.2f}, y = {y:.4f}")

# Problema (c)
ys_c, ts_c = taylor_order2(a=1.0, b=2.0, f=problem_c, f_second=problem_c_second, y_t0=2.0, h=0.25)
print("\nResultados para el problema (c):")
for t, y in zip(ts_c, ys_c):
    print(f"t = {t:.2f}, y = {y:.4f}")

# Problema (d)
ys_d, ts_d = taylor_order2(a=0.0, b=1.0, f=problem_d, f_second=problem_d_second, y_t0=1.0, h=0.25)
print("\nResultados para el problema (d):")
for t, y in zip(ts_d, ys_d):
    print(f"t = {t:.2f}, y = {y:.4f}")
```
